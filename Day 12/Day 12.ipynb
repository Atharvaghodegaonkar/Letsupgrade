```python
import numpy as np
import pandas as pd
from scipy.stats import wilcoxon, friedmanchisquare, mannwhitneyu, chi2_contingency , kruskal
```

```python
df = pd.read_csv('general_data.csv')
df
```

```python
df.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Attrition'] = encoder.fit_transform(df['Attrition'])
```

* We can use chi-square test to check dependace between two categorical value
* H0 : There is no dependancy between Attrition and BusinessTravel
* H1 : There is dependancy between Attrition and BusinessTravel

```python
df.columns
```

```python
chitable = pd.crosstab(df['Attrition'] , df['BusinessTravel'])
chitable
```

```python
stat, p, dof, expected = chi2_contingency(chitable)
if p < 0.5 :
    print('There is dependancy between Attrition and BusinessTravel')
else :
    print('There is no dependancy between Attrition and BusinessTravel')
      
```

```python
chitable = pd.crosstab(df['Attrition'] , df['Department'])
chitable
```

```python
stat, p, dof, expected = chi2_contingency(chitable)
if p < 0.5 :
    print('There is dependancy between Attrition and Department')
else :
    print('There is no dependancy between Attrition and Department')
```
