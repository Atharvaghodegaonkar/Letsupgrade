### For all the condition check value of alpha is 0.5

* We will check all the continues features and there relation between attrition 
* If there is relation then print r value and if there is no relation print the statement
* For all the cases the 
* H0: There is no relation
* H1 : There is realtion between two variable

```python
list_two_check = ['Age' , 'DistanceFromHome' , 'MonthlyIncome' , 'PercentSalaryHike' , 'MonthlyIncome' , 'TrainingTimesLastYear' , 'YearsAtCompany', 'YearsSinceLastPromotion' , 'YearsWithCurrManager']
for i in list_two_check :
    stats,P = pearsonr(df['Attrition'] , df[i])
    if P >= 0.05:
        print('There is no corelation between attrition and {}'.format(i))
    else :
        if stats < 0 :
            print('There is negitive relation between {} and {} and r is {}'.format('Attrition' , i , stats))
        else :
            print('There is positive relation between {} and {} and r is {}'.format('Attrition' , i , stats))
            
```
